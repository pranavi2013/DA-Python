```python
data_numeric=data.select_dtypes(include=['number'])
def find_IQR(col):
  q1=np.percentile(data_numeric,25)
  q3=np.percentile(data_numeric,75)
  iqr=q3-q1
  lowerbound=iqr-1.5*q1
  upperbound=iqr+1.5*q3
  return (lowerbound,upperbound)
for i in data_numeric.columns:
  print(find_IQR(i))
```
